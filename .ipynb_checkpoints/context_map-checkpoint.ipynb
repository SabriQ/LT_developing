```python
import os,sys,glob,re
import pandas as pd
import numpy as np
from mylab.ana.miniscope.context_exposure.Canamini import build_session
```

```python
def generate_columns(session):
    mouse_id = re.findall("(\d+)_part",session)[0]
    part = re.findall("part(\d+)",session)[0]

    
    s = build_session(session)
    if not s.exp == "hc":
        aim = s.exp
        index=s.result["index"][0]
    else:
        aim="hc"
        index="00000000-000000"
    
    return mouse_id,part,aim,index
```


# all sessions(new)

```python
sessions = glob.glob(r"\\10.10.47.163\Data_archive\qiushou\Sessions\*.pkl")
```

# save as csv

```python
mouse_id=[]
part=[]
aim=[]
index=[]
for session in sessions:
    mi,pa,ai,ind = generate_columns(session)
    print(ind)
    mouse_id.append(mi)
    part.append(pa)
    aim.append(ai)
    index.append(ind)


df = pd.DataFrame()
df["mouse_id"] = mouse_id
df["part"] = part
df["aim"]=aim
df["index"]=index

df.to_csv(r"C:\Users\qiushou\Desktop\context_map.csv")
```
